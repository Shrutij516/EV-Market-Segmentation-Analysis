```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("C:/Users/bhavy/Downloads/electric_vehicle_charging_station_list.csv")
```

```python
# Overview of the dataset
print(df.head())

# Summary statistics
print(df.describe())
```

```python
# Count by region
region_counts = df['region'].value_counts()
print(region_counts)
```

```python
# Count by type
type_counts = df['type'].value_counts()
print(type_counts)
```

```python
# Count by power
power_counts = df['power'].value_counts()
print(power_counts)
```

```python
# Create a bar plot of region counts
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='region')
plt.title('Electric Vehicle Charging Points by Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Create a pie chart of type counts
plt.figure(figsize=(8, 8))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Electric Vehicle Charging Point Types')
plt.axis('equal')
plt.show()
```

```python
# Create a pie chart of charging point services
service_counts = df['service'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(service_counts, labels=service_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Charging Point Services')
plt.axis('equal')
plt.show()
```

```python
# Create a box plot of charging point power levels
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, y='power')
plt.title('Distribution of Charging Point Power Levels')
plt.ylabel('Power')
plt.show()
```
