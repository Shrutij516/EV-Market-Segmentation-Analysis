```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the data from the CSV file into a DataFrame
df = pd.read_csv("C:/Users/bhavy/Downloads/Indian automoble buying behavour study 1.0.csv")
```

```python
# Analysis 1: Age Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', bins=10)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Analysis 2: Profession Analysis
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Profession')
plt.title('Profession Analysis')
plt.xlabel('Profession')
plt.ylabel('Count')
plt.show()
```

```python
# Analysis 3: Marital Status Analysis
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Marrital Status')
plt.title('Marital Status Analysis')
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.show()
```

```python
# Analysis 4: Education Analysis
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Education')
plt.title('Education Analysis')
plt.xlabel('Education')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Analysis 5: Total Salary Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Total Salary', bins=10)
plt.title('Total Salary Distribution')
plt.xlabel('Total Salary')
plt.ylabel('Count')
plt.show()
```

```python
# Analysis 6: Vehicle Make Analysis
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Make')
plt.title('Vehicle Make Analysis')
plt.xlabel('Make')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate the car spending percentage
df['Car Spending %'] = (df['Price'] / df['Total Salary']) * 100

# Identify and remove the outlier
outlier_index = df['Car Spending %'].idxmax()  # Assuming the outlier is the maximum car spending percentage
df = df.drop(outlier_index)

# Plot the car spending percentage
plt.figure(figsize=(20, 6))
plt.bar(df.index, df['Car Spending %'])
plt.title('Car Spending Percentage')
plt.xlabel('Purchase Index')
plt.ylabel('Car Spending %')
plt.xticks(df.index, df['Make'], rotation=90)
plt.show()
```

```python
# Group the data by marital status and calculate average salary and average car price for each group
grouped_df = df.groupby('Marrital Status').agg({'Salary': 'mean', 'Price': 'mean'})

# Plot the average salary and average car price for each marital status
plt.figure(figsize=(10, 6))
grouped_df.plot(kind='bar')
plt.title('Average Salary and Car Price by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Amount')
plt.legend(['Average Salary', 'Average Car Price'])
plt.xticks(rotation=0)
plt.show()
```

```python
# Create a figure and axes for each factor
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

# Function to format y-axis labels as salary amounts with rupee symbol
def salary_formatter(x, pos):
    return '₹{:,.0f}'.format(x)

# Group the data by marital status and plot salary and car price for each group
grouped_marital = df.groupby('Marrital Status').agg({'Salary': 'mean', 'Price': 'mean'})
grouped_marital.plot(kind='bar', ax=axes[0, 0])
axes[0, 0].set_title('Average Salary and Car Price by Marital Status')
axes[0, 0].set_xlabel('Marital Status')
axes[0, 0].set_ylabel('Amount')
axes[0, 0].yaxis.set_major_formatter(FuncFormatter(salary_formatter))

# Group the data by education and plot salary and car price for each group
grouped_education = df.groupby('Education').agg({'Salary': 'mean', 'Price': 'mean'})
grouped_education.plot(kind='bar', ax=axes[0, 1])
axes[0, 1].set_title('Average Salary and Car Price by Education')
axes[0, 1].set_xlabel('Education')
axes[0, 1].set_ylabel('Amount')
axes[0, 1].yaxis.set_major_formatter(FuncFormatter(salary_formatter))

# Group the data by profession and plot salary and car price for each group
grouped_profession = df.groupby('Profession').agg({'Salary': 'mean', 'Price': 'mean'})
grouped_profession.plot(kind='bar', ax=axes[1, 0])
axes[1, 0].set_title('Average Salary and Car Price by Profession')
axes[1, 0].set_xlabel('Profession')
axes[1, 0].set_ylabel('Amount')
axes[1, 0].yaxis.set_major_formatter(FuncFormatter(salary_formatter))

# Group the data by age and plot salary and car price for each group
grouped_age = df.groupby('Age').agg({'Salary': 'mean', 'Price': 'mean'})
grouped_age.plot(kind='bar', ax=axes[1, 1])
axes[1, 1].set_title('Average Salary and Car Price by Age')
axes[1, 1].set_xlabel('Age')
axes[1, 1].set_ylabel('Amount')
axes[1, 1].yaxis.set_major_formatter(FuncFormatter(salary_formatter))

# Adjust the layout and spacing between subplots
plt.tight_layout()

# Display the plots
plt.show()
```
